AmaÃ§: **Fiyat Elastikiyetini** (Price Elasticity) bulmak.

```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('../data/mock_sales_data.csv')
df.head()
```

## 1. Veri HazÄ±rlÄ±ÄŸÄ±
Basitlik adÄ±na tek bir Ã¼rÃ¼ne odaklanalÄ±m.

```python
top_product = df.groupby('product_id')['quantity'].sum().idxmax()
print(f"Analiz Edilecek ÃœrÃ¼n: {top_product}")

prod_data = df[df['product_id'] == top_product].copy()

# Fiyat vs Miktar iliÅŸkisi
sns.scatterplot(data=prod_data, x='unit_price', y='quantity')
plt.title(f'Fiyat - Talep EÄŸrisi ({top_product})')
plt.show()
```

## 2. Log-Log Regresyon (Elastikiyet HesabÄ±)
Ä°ktisat teorisine gÃ¶re: `ln(Q) = a + b * ln(P)`
Buradaki `b` katsayÄ±sÄ± bize elastikiyeti verir.

```python
prod_data['ln_price'] = np.log(prod_data['unit_price'])
prod_data['ln_quantity'] = np.log(prod_data['quantity'])

X = prod_data['ln_price']
y = prod_data['ln_quantity']
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

- EÄŸer **-1 ile 0 arasÄ±ndaysa**, talep duyarsÄ±zdÄ±r (Ä°nelastik).
