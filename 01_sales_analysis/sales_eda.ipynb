```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ayarlar
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Veri YÃ¼kleme
Mock verimizi yÃ¼klÃ¼yoruz.

```python
df = pd.read_csv('../data/mock_sales_data.csv')
df['date'] = pd.to_datetime(df['date'])
df.head()
```

## 2. Genel BakÄ±ÅŸ
Toplam ciro, satÄ±ÅŸ adedi ve kategori bazlÄ± performans.

```python
df['revenue'] = df['quantity'] * df['unit_price']
print(f"Toplam Ciro: â‚º{df['revenue'].sum():,.2f}")
print(f"Toplam Ä°ÅŸlem SayÄ±sÄ±: {len(df):,}")
```

## 3. Zaman Ä°Ã§indeki SatÄ±ÅŸ Trendi
HaftalÄ±k trendleri inceleyelim.

```python
daily_sales = df.groupby('date')['revenue'].sum().reset_index()
sns.lineplot(data=daily_sales, x='date', y='revenue')
plt.title('GÃ¼nlÃ¼k SatÄ±ÅŸ Trendi')
plt.show()
```

## 4. Kategori PerformansÄ±
Hangi kategori ciroyu sÄ±rtlÄ±yor?

```python
cat_summary = df.groupby('category')['revenue'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=cat_summary, x='revenue', y='category', palette='viridis')
plt.title('Kategori BazlÄ± Toplam Ciro')
plt.show()
```
