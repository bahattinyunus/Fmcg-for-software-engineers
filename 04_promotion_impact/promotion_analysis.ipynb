# ðŸ“£ Promotion Impact Analysis

Bu analiz, indirimli (Promo) satÄ±ÅŸlarÄ±n normal satÄ±ÅŸlara kÄ±yasla ne kadar **Uplift** (artÄ±ÅŸ) yarattÄ±ÄŸÄ±nÄ± Ã¶lÃ§er.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv('../data/mock_sales_data.csv')
df.head()
```

## 1. Promo vs Non-Promo KarÅŸÄ±laÅŸtÄ±rmasÄ±
Ortalama satÄ±ÅŸ adetlerini karÅŸÄ±laÅŸtÄ±ralÄ±m.

```python
promo_stats = df.groupby('is_promo')['quantity'].mean().reset_index()
print(promo_stats)

sns.barplot(data=promo_stats, x='is_promo', y='quantity')
plt.title('Promo Etkisi: Ortalama Adet SatÄ±ÅŸÄ±')
plt.show()
```

## 2. Ä°statistiksel Test (T-Test)
Aradaki fark istatistiksel olarak anlamlÄ± mÄ±?

```python
promo_sales = df[df['is_promo'] == True]['quantity']
regular_sales = df[df['is_promo'] == False]['quantity']

t_stat, p_val = stats.ttest_ind(promo_sales, regular_sales)
print(f"T-Statistic: {t_stat:.2f}, P-Value: {p_val:.4f}")

if p_val < 0.05:
    print("SonuÃ§: Promosyonun satÄ±ÅŸlar Ã¼zerinde ANLAMLI bir pozitif etkisi var.")
else:
    print("SonuÃ§: Promosyon etkisi anlamsÄ±z (veya veri yetersiz).")
```
