# üîÆ Demand Forecasting (Talep Tahmini)

Bu notebook, **Prophet** k√ºt√ºphanesini kullanarak gelecekteki satƒ±≈ü adetlerini tahminler.

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Veri Y√ºkleme
df = pd.read_csv('../data/mock_sales_data.csv')
df['date'] = pd.to_datetime(df['date'])

# G√ºnl√ºk Toplam Satƒ±≈ü Adedi (Quantity) hesaplama
daily_sales = df.groupby('date')['quantity'].sum().reset_index()
daily_sales.columns = ['ds', 'y'] # Prophet formatƒ±
daily_sales.head()
```

## 1. Modelleme
Facebook Prophet modelini kurup eƒüitiyoruz.

```python
m = Prophet(daily_seasonality=True, weekly_seasonality=True)
m.fit(daily_sales)

# Gelecek 30 g√ºn i√ßin tahmin
future = m.make_future_dataframe(periods=30)
forecast = m.predict(future)

forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

## 2. G√∂rselle≈ütirme
Ger√ßekle≈üen vs Tahmin.

```python
fig1 = m.plot(forecast)
plt.title('Gelecek 30 G√ºnl√ºk Satƒ±≈ü Tahmini')
plt.show()
```

## 3. Trend ve Mevsimsellik Bile≈üenleri
Haftanƒ±n hangi g√ºn√º satƒ±≈ülar artƒ±yor?

```python
fig2 = m.plot_components(forecast)
```
